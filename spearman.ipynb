```python
import pandas as pd
def read_all_df(file_suffix=".jsonl"):
    # read all df with file prefix in this directory
    import os
    dfs = []
    for file in os.listdir()[:10]:
        if file.endswith(file_suffix):
            print(file)
            tf,ef = file.split('_on_')
            evaluation = ef
            train = tf
            df = pd.read_json(file,lines=True)
            df["train"] = train
            df["evaluation"] = evaluation.split(".")[0]

            dfs.append(df)
    df = pd.concat(dfs, ignore_index=True)
    df = df.reset_index()
    pre_melt_df = df
    return pre_melt_df


pre_melt_df = read_all_df()
pre_melt_df.head()
```

# Calculate spearmans

```python
from scipy import stats
import pandas as pd

def generate_compute_spearman(prop1,prop2):
    def _compute_spearman(row):
        try:
            corr, _ = stats.spearmanr(row[prop1], row[prop2])
            return corr
        except:
            return None
    return _compute_spearman


df = pre_melt_df
for iteration in range(1):

    df[f"degrees_{iteration}"] = df['degrees'].apply(lambda x:  x[iteration] if iteration<len(x) else None)
    df[f"probs_{iteration}"] = df['probs'].apply(lambda x: x[iteration] if iteration<len(x) else None)
    df[f"pc1_{iteration}_layer_0"] = df['pca_layer_0'].apply(lambda x: x[iteration] if iteration<len(x) else None)
    df[f"pc1_{iteration}_layer_1"] = df['pca_layer_1'].apply(lambda x: x[iteration] if iteration<len(x) else None)
    df[f"pc1_{iteration}_layer_2"] = df['pca_layer_2'].apply(lambda x: x[iteration] if iteration<len(x) else None)
    df[f"pc1_{iteration}_layer_3"] = df['pca_layer_3'].apply(lambda x: x[iteration] if iteration<len(x) else None)

    df[f"spearman_between_degrees_and_probs_iter_{iteration}"] = df.apply(generate_compute_spearman(f"degrees_{iteration}",f"probs_{iteration}"), axis=1)

    df[f"spearman_between_degrees_and_pc1_layer_0_iter_{iteration}"] = df.apply(generate_compute_spearman(f"degrees_{iteration}",f"pc1_{iteration}_layer_0"), axis=1)
    df[f"spearman_between_degrees_and_pc1_layer_1_iter_{iteration}"] = df.apply(generate_compute_spearman(f"degrees_{iteration}",f"pc1_{iteration}_layer_1"), axis=1)
    df[f"spearman_between_degrees_and_pc1_layer_2_iter_{iteration}"] = df.apply(generate_compute_spearman(f"degrees_{iteration}",f"pc1_{iteration}_layer_2"), axis=1)
    df[f"spearman_between_degrees_and_pc1_layer_3_iter_{iteration}"] = df.apply(generate_compute_spearman(f"degrees_{iteration}",f"pc1_{iteration}_layer_3"), axis=1)

    df[f"spearman_between_probs_and_pc1_layer_0_iter_{iteration}"] = df.apply(generate_compute_spearman(f"probs_{iteration}",f"pc1_{iteration}_layer_0"), axis=1)
    df[f"spearman_between_probs_and_pc1_layer_1_iter_{iteration}"] = df.apply(generate_compute_spearman(f"probs_{iteration}",f"pc1_{iteration}_layer_1"), axis=1)
    df[f"spearman_between_probs_and_pc1_layer_2_iter_{iteration}"] = df.apply(generate_compute_spearman(f"probs_{iteration}",f"pc1_{iteration}_layer_2"), axis=1)
    df[f"spearman_between_probs_and_pc1_layer_3_iter_{iteration}"] = df.apply(generate_compute_spearman(f"probs_{iteration}",f"pc1_{iteration}_layer_3"), axis=1)

```

# Q0 Degree 10 iterations

```python
def unmask(mask, smaller_array):
    """
    smaller_array might not be smaller :p
    """
    ret = [_ for _ in mask]
    smaller_index = 0
    for i, item in enumerate(mask):
        if i >= len(smaller_array):
            ret[i] = None
        elif item:
            ret[i] = smaller_array[smaller_index]
            smaller_index += 1
        else:
            ret[i] = None
    return ret
```

```python
import numpy as np
columns_to_show_degrees = []
def clique_members_to_mask(row):
    clique_members_mask = np.zeros(row['graph_size'], dtype=bool)
    clique_members_mask[row['clique']] = True
    return clique_members_mask

def to_quantiles(arr):
    if arr is None:
        return None
    return np.digitize(
            arr, bins=np.quantile(arr, [0.25, 0.5, 0.75]), right=False
        )

df["graph_size"] = df["graph_sizes"].apply(lambda x:x[0])
df["non_clique_size"] = df["graph_size"] - df["clique_size"]
df["clique_mask"] = df.apply(clique_members_to_mask,axis=1)
for iteration in range(11):
    def how_many_in_degree_q0(row):
        quantiles = to_quantiles(row[f"iter_{iteration}_q0_degrees"])
        if quantiles is None:
            return None
        quantiles_clique = np.array(unmask(row["clique_mask"],quantiles))
        return sum(quantiles_clique==0)/row["clique_size"]
    def how_many_non_clique_in_q0(row):
        quantiles = to_quantiles(row[f"iter_{iteration}_q0_degrees"])
        if quantiles is None:
            return None
        quantiles_clique = np.array(unmask(row["clique_mask"],quantiles))
        return (sum(quantiles==0)-sum(quantiles_clique==0))/row["graph_size"]
    df[f"iter_{iteration}_q0_degrees"] = df['degrees'].apply(lambda x: x[iteration] if iteration<len(x) else None)
    df[f"iter_{iteration}_degrees_clique_in_q0"] = df.apply(how_many_in_degree_q0,axis=1)
    df[f"iter_{iteration}_degrees_non_clique_in_q0"] = df.apply(how_many_non_clique_in_q0,axis=1)

df[f"iter_0_degrees_clique_in_q0"]
# df[columns_to_show_degrees]
```

```python
columns_to_show_degrees = []
degrees_clique_columns = [f"iter_{i}_degrees_clique_in_q0" for i in range(11)]
df[f"degrees_clique_sum"] = df[degrees_clique_columns].sum(axis=1)
df[f"degrees_clique_count"] = (~df[degrees_clique_columns].isna()).sum(axis=1)
df[f"degrees_clique"] = df[f"degrees_clique_sum"]/df[f"degrees_clique_count"]
columns_to_show_degrees.append(f"degrees_clique")

degrees_non_clique_columns = [f"iter_{iteration}_degrees_non_clique_in_q0" for i in range(11)]
df[f"degrees_non_clique_sum"] = df[degrees_non_clique_columns].sum(axis=1)
df[f"degrees_non_clique_count"] = (~df[degrees_non_clique_columns].isna()).sum(axis=1)
df[f"degrees_non_clique"] = df[f"degrees_non_clique_sum"]/df[f"degrees_non_clique_count"]
columns_to_show_degrees.append(f"degrees_non_clique")
```

# Q0 Probs 10 iterations

```python
import numpy as np
columns_to_show_probs = []
def clique_members_to_mask(row):
    clique_members_mask = np.zeros(row['graph_size'], dtype=bool)
    clique_members_mask[row['clique']] = True
    return clique_members_mask

def to_quantiles(arr):
    if arr is None:
        return None
    return np.digitize(
            arr, bins=np.quantile(arr, [0.25, 0.5, 0.75]), right=False
        )

df["graph_size"] = df["graph_sizes"].apply(lambda x:x[0])
df["non_clique_size"] = df["graph_size"] - df["clique_size"]
df["clique_mask"] = df.apply(clique_members_to_mask,axis=1)
for iteration in range(11):
    def how_many_in_degree_q0(row):
        if row[f"iter_{iteration}_probs"] is None or all(np.array(row[f"iter_{iteration}_probs"])==None):
            return None
        quantiles = to_quantiles(row[f"iter_{iteration}_probs"])
        if quantiles is None:
            return None
        quantiles_clique = np.array(unmask(row["clique_mask"],quantiles))
        return sum(quantiles_clique==0)/row["clique_size"]
    def how_many_non_clique_in_q0(row):
        if row[f"iter_{iteration}_probs"] is None or all(np.array(row[f"iter_{iteration}_probs"])==None):
            return None
        quantiles = to_quantiles(row[f"iter_{iteration}_probs"])
        if quantiles is None:
            return None
        quantiles_clique = np.array(unmask(row["clique_mask"],quantiles))
        return (sum(quantiles==0)-sum(quantiles_clique==0))/row["graph_size"]
    df[f"iter_{iteration}_probs"] = df['probs'].apply(lambda x: x[iteration] if iteration<len(x) else None)
    df[f"iter_{iteration}_probs_clique_in_q0"] = df.apply(how_many_in_degree_q0,axis=1)
    df[f"iter_{iteration}_probs_non_clique_in_q0"] = df.apply(how_many_non_clique_in_q0,axis=1)

df[f"iter_0_probs_clique_in_q0"]
# df[columns_to_show_probs]
```

```python
columns_to_show_probs = []
probs_clique_columns = [f"iter_{i}_probs_clique_in_q0" for i in range(11)]
df[f"probs_clique_sum"] = df[probs_clique_columns].sum(axis=1)
df[f"probs_clique_count"] = (~df[probs_clique_columns].isna()).sum(axis=1)
df[f"probs_clique"] = df[f"probs_clique_sum"]/df[f"probs_clique_count"]
columns_to_show_probs.append(f"probs_clique")

probs_non_clique_columns = [f"iter_{i}_probs_non_clique_in_q0" for i in range(11)]
df[f"probs_non_clique_sum"] = df[probs_non_clique_columns].sum(axis=1)
df[f"probs_non_clique_count"] = (~df[probs_non_clique_columns].isna()).sum(axis=1)
df[f"probs_non_clique"] = df[f"probs_non_clique_sum"]/df[f"probs_non_clique_count"]
columns_to_show_probs.append(f"probs_non_clique")
```
